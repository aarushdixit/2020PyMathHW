## Writing a Function in Python using `def` and `return`

```python
# Define a function
# Keywords: def and return
# Function name: my_square_root

def my_square_root(x): 
    return np.sqrt(x)
```

## Rose Curves: A Family of Functions

```python
# Integer parameter k
k = 2

t = np.linspace(0,4*np.pi,1000)

x = np.cos(k*t) * np.cos(t)
y = np.cos(k*t) * np.sin(t)

plt.plot(x,y,color = 'b',linewidth=5)
plt.axis('square')
plt.axis('off')
plt.show()
```

### Rose Curve with Decimal Parameter

```python
# Decimal parameter value
k = 4.2

t = np.linspace(0,8*np.pi,1000)

x = np.cos(k*t) * np.cos(t)
y = np.cos(k*t) * np.sin(t)

plt.plot(x,y,color = 'm',linewidth=5)
plt.axis('square')
plt.axis('off')
plt.show()
```

### Plot Rose Curves on a 5-by-5 Grid using `plt.subplots`

```python
# Grid: 3 rows and 3 columns

t = np.linspace(0,8*np.pi,1000)

fig,ax = plt.subplots(3,3,figsize=(10,6))
row, col = np.indices((3, 3))

print ("k row col")
for i in range(0, 9):
    
    k = i/2
    
    
    x = np.cos(k*t) * np.cos(t)
    y = np.cos(k*t) * np.sin(t)
    
    r = row.ravel()[i]
    c = col.ravel()[i]
    
    print (k, r, c)
    
    ax[r,c].plot(x,y,color = 'm', linewidth=2)
    ax[r,c].axis('off')
    ax[r,c].axis('square')
    ax[r,c].set_title('k=%s'%k)
    
plt.show()
```

### Homework: Logarithmic spiral

+ Read the Wikipedia Page about the logarithmic Spirals:
    - https://en.wikipedia.org/wiki/Logarithmic_spiral
+ Using the example code below, investigate how logarithmic spirals change for different parameter values

```python
# Logarithmic spiral for parameter k = -3
t = np.linspace(0,10*np.pi,1000)

k = -3

x = np.cos(t) * np.exp(t/k)
y = np.sin(t) * np.exp(t/k)

plt.plot(x,y,'m',linewidth=5)
plt.axis('square')
plt.axis('off')
plt.show()
```

### Exercise
+ Generate N = 150 spirals
+ Vary the color of each spiral to be defined in terms of RGB (Red, Green, Blue) as `color=[i/N,0,i/N]`

```python
# Green Rose Spiral
N = 150

# generate N different values for parameter k
# parameter k: from -6 to -2 in equally spaced intervals
ks = np.linspace(-6,-2,N)

for i in range(0,N):
    x = np.cos(t) * np.exp(t/ks[i])
    y = np.sin(t) * np.exp(t/ks[i])
    plt.plot(x,y,color="g")
    
plt.axis('square')
plt.axis('off')
plt.show()
```
